```python
# Dependencies and Setup
import pandas as pd
import numpy as np

```

```python
# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"
# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

## Player Count

* Display the total number of players


```python
total_num_players_list = purchase_data["SN"].value_counts()
total_num_players_list
```

```python
total_num_players = len(total_num_players_list)
total_num_players
```

```python
total_player_count = pd.DataFrame({"Total Players":[total_num_players]})
total_player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#For the number of unique items
no_unique_items = len((purchase_data['Item ID']).unique())
no_unique_items
```

```python
#for average price
avg_price= purchase_data['Price'].mean()
avg_price
```

```python
#for number of purchases
no_of_purchases = len((purchase_data['Purchase ID'].unique()))
no_of_purchases
```

```python
#for Total Revenue
total_rev = purchase_data['Price'].sum()
total_rev
```

```python
results_df = pd.DataFrame({"Number of Unique Items":[no_unique_items], "Average Price":[avg_price],
                          "Number of purchases":[no_of_purchases], "Total Revenue":[total_rev]})

results_df.style.format({'Average Price':"${:,.2f}",
                        'Total Revenue':"${:,.2f}"})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#for binning the age
bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 80]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
```

```python
purchase_data['Age Group'] = pd.cut(purchase_data['Age'], bins, labels=group_names)
purchase_data
```

```python
grouped_age = purchase_data.groupby("Age Group")
#for total count
age_total_count = grouped_age["SN"].nunique()

#for Percentage
age_percentage = (age_total_count/total_num_players) * 100
```

```python
#for table
age_table = pd.DataFrame({"Total Count": age_total_count, 
                          "Percentage of Players": age_percentage})
age_table
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Creating a new table for items
items = items = purchase_data[["Item ID", "Item Name", "Price"]]
items
```

```python
#To group them
item_group = items.groupby(["Item ID", "Item Name"])

#To count them by price
item_count = item_group["Price"].count()
item_count
```

```python
#To get purchase value
item_value = item_group["Price"].sum()
item_value
```

```python
#To get each item price
item_price = item_value/item_count
item_price
```

```python
#for the table
item_table = pd.DataFrame({"Purchase Count": item_count,
                          "Item Price": item_price,
                          "Total Purchase Value": item_value})

item_table.style.format({"Item Price":"${:,.2f}",
                            "Total Purchase Value":"${:,.2f}"})
```

```python
new_item_table = item_table.sort_values(by='Purchase Count', ascending=False).head()
new_item_table.style.format({"Item Price":"${:,.2f}",
                            "Total Purchase Value":"${:,.2f}"})
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
profitable_items = new_item_table.sort_values(by="Total Purchase Value", ascending=False)
profitable_items.style.format({"Item Price":"${:,.2f}",
                            "Total Purchase Value":"${:,.2f}"})
```
